```python
## Importing the libraries
from numpy import genfromtxt
from time import time
from datetime import datetime
from sqlalchemy import Column, Integer, Float, Date, String, VARCHAR
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import create_engine
from sqlalchemy.orm import sessionmaker
import csv
import pandas as pd
import sys
import json
import os


## Defining tables of the database
Base = declarative_base()

class customer(Base):
    #Tell SQLAlchemy what the table name is and if there's any table-specific arguments it should know about
    __tablename__ = 'Customer_details'
    __table_args__ = {'sqlite_autoincrement': True}
    #tell SQLAlchemy the name of column and its attributes:
    customer_id = Column(VARCHAR(51), primary_key=True, nullable=False)
    customer_zip_code_prefix = Column(Integer)
    customer_city = Column(VARCHAR(31))
    customer_state = Column(VARCHAR(11))
fields1 = ['customer_id','customer_zip_code_prefix','customer_city','customer_state']

class hits(Base):
    #Tell SQLAlchemy what the table name is and if there's any table-specific arguments it should know about
    __tablename__ = 'Hits_counter'
    __table_args__ = {'sqlite_autoincrement': True}
    #tell SQLAlchemy the name of column and its attributes:
    product_id = Column(VARCHAR(51), primary_key=True, nullable=False)
    product_hits = Column(Integer)     
fields2 = ['product_id','product_hits']

class payment(Base):
    #Tell SQLAlchemy what the table name is and if there's any table-specific arguments it should know about
    __tablename__ = 'Payment'
    __table_args__ = {'sqlite_autoincrement': True}
    #tell SQLAlchemy the name of column and its attributes:
    customer_id = Column(VARCHAR(51))
    product_id = Column(VARCHAR(51))
    order_id = Column(VARCHAR(51), primary_key=True, nullable=False)
    payment_type = Column(VARCHAR(51))
    payment_installments = Column(Integer)
    payment_value = Column(Float)       
fields3 = ['customer_id','product_id','order_id','payment_type','payment_installments','payment_value']

class product(Base):
    #Tell SQLAlchemy what the table name is and if there's any table-specific arguments it should know about
    __tablename__ = 'Product_catalogue'
    __table_args__ = {'sqlite_autoincrement': True}
    #tell SQLAlchemy the name of column and its attributes:
    product_id = Column(VARCHAR(51), primary_key=True, nullable=False)
    product_url = Column(VARCHAR(251))
    product_name = Column(VARCHAR(151))
    product_category_tree = Column(VARCHAR(255))
    retail_price = Column(Integer)
    discounted_price = Column(Integer)
    product_rating = Column(Integer)
    brand = Column(VARCHAR(31))
fields4 = ['product_id','product_url','product_name','product_category_tree','retail_price','discounted_price','product_rating','brand']



## Importing data from csv files
file_name1 = '_customers_dataset.csv'
df1 = pd.read_csv(file_name1, usecols = fields1)

file_name2 = '_hits_counter.csv'
df2 = pd.read_csv(file_name2, usecols = fields2)

file_name3 = '_order_payments_dataset.csv'
df3 = pd.read_csv(file_name3, usecols = fields3)

file_name4 = '_product_catalogue.csv'
df4 = pd.read_csv(file_name4, usecols = fields4)


## Creating the database
engine = create_engine('sqlite:///Keval_SQLite.db')
Base.metadata.create_all(engine)

df1.to_sql(con=engine, index_label='id', name=customer.__tablename__, if_exists='replace')
df2.to_sql(con=engine, index_label='id', name=hits.__tablename__, if_exists='replace')
df3.to_sql(con=engine, index_label='id', name=payment.__tablename__, if_exists='replace')
df4.to_sql(con=engine, index_label='id', name=product.__tablename__, if_exists='replace')
```
